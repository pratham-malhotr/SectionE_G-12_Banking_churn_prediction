# Notebook 01: Data Extraction

In this notebook, we load the raw data and prove that it passes the **Capstone 2 Minimum Standards** (At least 5,000 rows and 8 meaningful analytical columns). We are not altering or cleaning any data here.

### Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
```

### Step 2: Load the Raw Dataset
Loading the completely untouched raw data from our `data/raw/` directory.

```python
# Read the CSV file into a pandas dataframe
df = pd.read_csv("../data/raw/Banking_churn_prediction.csv")

# Show the first 5 rows to confirm it loaded correctly
df.head()
```

### Step 3: Check Capstone 2 Requirements
We need to prove to the evaluator that our dataset is large enough! Minimum Requirements: 5,000 rows and 8 columns.

```python
num_rows = df.shape[0]
num_cols = df.shape[1]

print("--- Capstone 2 Minimum Standards Check ---")

# Check row count
if num_rows >= 5000:
    print("Passed: Dataset has at least 5000 rows. (Total Rows:", num_rows, ")")
else:
    print("Failed: Dataset is too small. (Total Rows:", num_rows, ")")
    
# Check column count
if num_cols >= 8:
    print("Passed: Dataset has at least 8 columns. (Total Columns:", num_cols, ")")
else:
    print("Failed: Not enough columns. (Total Columns:", num_cols, ")")
```
